```python
import pandas as pd
import numpy as np
import matplotlib as plt
from numpy.random import randn
%matplotlib inline

# Create sample dataframe
my_df = pd.DataFrame(randn(100, 5), columns=["Mon", "Tue", "Wed", "Thu", "Fri"])
my_df
```

```python
#Histogram
my_df.hist("Wed")
```

```python
# Histograms with smaller bins
my_df["Wed"].hist(bins=40)
```

```python
my_df["Wed"].plot(kind="hist")
```

```python
my_df["Wed"].hist(bins=50, grid= False, legend= True)
```

```python
# Area plots
my_df.plot(kind="area", stacked= False)        # For stacked= True, all values must be +ve, we have -ve values too
```

```python
# Taking absolute values of the negative values
my_df.abs().plot(kind="area")
```

```python
# Changing colors of the plot
# my_df.abs().plot(kind="area", alpha=1)   #Default
my_df.abs().plot(kind="area", alpha=0.39)         #alpha: 0-1
```

```python
# Adding title
my_df.abs().plot(kind="area", title= "My Area plot", alpha=0.5)
```

```python
# Removing legends
my_df.abs().plot(kind="area", title= "My Area plot", alpha=0.5, legend = False)
```

```python
# Subplot one column
my_df["Mon"].abs().plot(kind="area", title= "My Area plot", alpha=0.5, legend=True)
```

```python
# Subplot with more than 1 column
my_df[["Mon", "Fri"]].abs().plot(kind="area", title= "My Area subplot", alpha=0.5, legend = True, grid=True)
```

```python
# Barcharts
my_df.plot(kind="bar", grid=True, title="My bar graph")
```

```python
# Absolute value
my_df.abs().plot(kind="bar", title="My bar graph")
```

```python
# Stacked bar graph
my_df.abs().plot(kind="bar", title="My bar graph", stacked=True)
```

```python
# Lineplots with Matplotlib
my_df.plot(kind="line")

```

```python
my_df.abs().plot(kind="line", grid=True, title="My bar graph")
```

```python
my_df["Mon"].plot(kind="line", grid=True, title="My bar graph")
```

```python
my_df[["Mon", "Fri"]].plot(kind="line", grid=True, title="My line graph")
```

```python
# Line width
my_df["Mon"].plot(kind="line", grid=True, title="My line graph", lw=20)
```

```python
my_df[["Mon", "Fri"]].plot(kind="line", grid=True, title="My line graph", lw=0.5)
```

```python
#Size
my_df[["Mon", "Wed"]].plot(kind="line", grid=True, title="My line graph", lw=2, figsize=(10,10))
```

```python
#Creating a new dataframe, just for fun
my_df2 = pd.DataFrame(randn(250,5), columns=["Mon", "Tue", "Wed", "Thu", "Fri"])

# Scatterplots
#my_df2.plot(kind="scatter") # Gives error cz for scatterplot we need to specify x and y axes.
my_df2.plot(kind="scatter", x="Mon", y="Tue")
```

```python
# Plot between 3 column values
my_df2.plot(kind="scatter", x="Mon", y="Tue", c="Wed")
```

```python
# Changing color
my_df2.plot(kind="scatter", x="Mon", y="Tue", c="Wed", cmap="magma")
```

```python
# by Size
my_df2.plot(kind="scatter", x="Mon", y="Tue", s="Wed", title= "My Scatterplot bc")
```

```python
# More size play
my_df2.plot(kind="scatter", x="Mon", y="Tue", s=my_df2['Wed']*100, title= "My Scatterplot bc")
```

```python
# Color with size
my_df2.plot(kind="scatter", x="Mon", y="Tue", s=my_df2['Wed']*50, title= "My Scatterplot bc", alpha=0.5)
```

```python
#Change color
my_df.plot(kind="box", cmap="jet")
```

```python
my_df.plot(kind="box", cmap="jet", grid=True)
```
