```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

my_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Insurance data.csv')
my_df.head()
```

```python
# Plotting the file to see the data distribution
plt.scatter(my_df.Age, my_df.Bought_insurance, marker='+', color='red')
```

```python
my_df.shape
```

```python
# Splitting the dataset into a and b
a = my_df.Age
b = my_df.Bought_insurance
a.shape, b.shape
```

```python
# Splitting the dataset into train and test datasets
from sklearn.model_selection import train_test_split
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=88)
print("Training set: ", a_train.shape, b_train.shape)
print("Testing set: ", a_test.shape, b_test.shape)
```

```python
# Create regression model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
```

```python


# Convert Pandas Series to NumPy array
#a_train_array = a_train.to_numpy()

# Reshape the array
#a_train_reshaped = a_train_array.reshape(-1, 1)

# Training the model
lr.fit(a_train, b_train)
```

```python
# Predict values using testing data
b_pred = lr.predict(a_train)
print(b_pred.shape)
print(b_test.shape)
```
