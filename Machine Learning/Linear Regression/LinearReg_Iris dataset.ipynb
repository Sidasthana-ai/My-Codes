```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib Inline

from sklearn.datasets import load_iris
iris = load_iris()
print(iris)
```

```python
# Spilitting data into a, b datasets
a = iris.data
b = iris.target
print(a.shape)
print(b.shape)
```

```python
# Splitting the data into training and testing datasets
from sklearn.model_selection import train_test_split
a_train, a_test, b_train, b_test = train_test_split(a, b, train_size=0.8, random_state=42)
print("Training dataset: ",a_train.shape, b_train.shape)
print("Testing dataset: ",a_test.shape, b_test.shape)
```

```python
# Linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Training the model using training datasets
lr.fit(a_train, b_train)
```

```python
# Getting prediction values using testing dataset
b_pred = lr.predict(a_test)
print("Prediction data: ", b_pred.shape)
print("Testing data: ", b_test.shape)
```

```python
# Performance evaluation
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2 = r2_score(b_pred, b_test)
mse = mean_squared_error(b_pred, b_test)
mae = mean_absolute_error(b_pred, b_test)
intercept = lr.intercept_
print("Model r2 score: ", r2)
print("Model mean squared error score: ", mse)
print("Model mean absolute error score: ", mae)
print("Model y-intercept: ", intercept)
```

```python
# Plotting Predicted value vs. Absolute value
plt.scatter(b_pred, b_test)
plt.plot([b.min(), b.max()],[b.min(), b.max()], color="red")
plt.ylabel("Predcited values")
plt.xlabel("Actual values")
plt.title("Predicted values vs. Absolute values")
plt.show()
```

```python
# Residual plot
plt.scatter(b_pred-b_test, b_pred)
plt.ylabel("Residual values")
plt.xlabel("Predicted values")
plt.title("Residual plot: Residual values vs. Predicted values")
plt.show()
```
