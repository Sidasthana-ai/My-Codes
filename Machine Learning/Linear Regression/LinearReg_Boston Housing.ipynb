```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import fetch_california_housing

calihousing = fetch_california_housing()
print(calihousing)
```

```python
# Spltting the data into a and b datasets
a = calihousing.data
b = calihousing.target
print(a.shape)
print(b.shape)
```

```python
# Divide dataset a and b in train and test datasets 80-20 split
from sklearn.model_selection import train_test_split
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)
print("Training set shape: ", a_train.shape, b_train.shape)
print("Testing set shape: ", a_test.shape, b_test.shape)
```

```python
# Linear regression model

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Training the model with train datasets
lr.fit(a_train, b_train)
```

```python
# Predictions from testing set
b_pred = lr.predict(a_test)
print("Prediction dataset shape: ", b_pred.shape)
print("Testing dataset shape: ", b_test.shape)       # Shapes must be equal
```

```python
# Performance evaluation
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r2 = r2_score(b_pred, b_test)
mse = mean_squared_error(b_pred, b_test)
mae = mean_absolute_error(b_pred, b_test)
intercept = lr.intercept_

print("R2 score of Linear Regression model: ", r2)
print("Mean square error of Linear Regression model: ", mse)
print("Mean absolute error of Linear Regression model: ", mae)
print("Intercept value of Linear Regression model: ", intercept)
```

```python
# Plot: Predicted value vs. Actual value

plt.scatter(b_pred, b_test, alpha=0.3)
plt.plot([b.min(), b.max()], [b.min(), b.max()], color="red")
plt.xlabel("Actual value")
plt.ylabel("Precidted value")
plt.title("Scatterplot: Predicted value vs. Actual value")
plt.show()
```

```python
#Plotting Residuals vs. Predicted values
plt.scatter(b_pred-b_test, b_pred)
plt.ylabel("Residuals")
plt.xlabel("Predicted value")
plt.title("Residual plot")
plt.show()
```
